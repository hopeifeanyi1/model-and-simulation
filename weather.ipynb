```python
import numpy as np
import matplotlib.pyplot as plt

# Define states
states = ["Sunny", "Cloudy", "Rainy"]

# Define transition matrix
P = np.array([[0.7, 0.2, 0.1],
              [0.3, 0.5, 0.2],
              [0.4, 0.4, 0.2]])

# Number of days to simulate
num_days = 100

# Start from "Sunny" (index 0)
current_state = 0
# Store weather sequence
weather_sequence = [states[current_state]]

# Simulate the Markov chain
for _ in range(num_days - 1):
    current_state = np.random.choice([0, 1, 2], p=P[current_state])
    weather_sequence.append(states[current_state])

# Plot the results
#more comment
plt.figure(figsize=(10, 4))
plt.plot(weather_sequence, marker="o", linestyle="-")
plt.xticks(rotation=90)
plt.title("Simulated Weather Over 100 Days")
plt.xlabel("Days")
plt.ylabel("Weather State")
plt.show()
```
